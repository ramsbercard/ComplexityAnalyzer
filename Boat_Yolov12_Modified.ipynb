```python
# =========================================
# Modular Vision Pipeline Demonstration
# (NumPy-Compatible with Real Images, Analyzer-Ready)
# =========================================

import importlib
import cv2
import numpy as np
import glob
import vision_pipeline

# Reload in case of edits
importlib.reload(vision_pipeline)
from vision_pipeline import *

# -----------------------------
# Parameters (Explicit & Symbolic)
# -----------------------------
n = None       # total dataset size (determined after loading images)
b = 4          # batch size
h = 128        # image height
w = 128        # image width
c = 3          # number of classes/channels
L = 4          # number of backbone layers
p = 10         # proposals per image
d = c          # feature depth (matching channels)

# -----------------------------
# Load Real Images
# -----------------------------
image_paths = glob.glob("test_images/*.jpg")  # adjust folder path
n = len(image_paths)

# Load batch of images with explicit shape (b, h, w, c)
real_images = np.zeros((b, h, w, c), dtype=np.float32)
for batch_idx, path in enumerate(image_paths[:b]):
    img = cv2.imread(path)
    img = cv2.resize(img, (w, h))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for i in range(h):
        for j in range(w):
            for ch in range(c):
                real_images[batch_idx][i][j][ch] = img[i][j][ch]

print(f"Loaded {len(real_images)} real images with shape: {real_images.shape}")

# -----------------------------
# Preprocessing
# -----------------------------
# Use fully analyzer-compatible preprocessing
processed = preprocessing(real_images, b, h, w, c)

# -----------------------------
# Forward Pass (Backbone)
# -----------------------------
features = forward_pass_detection(processed, b, L, h, w, d)

# -----------------------------
# Detection Head
# -----------------------------
detections = detection_head(features, b, p, c)

# -----------------------------
# Segmentation Head
# -----------------------------
masks = segmentation_head(features, b, h, w, c)

# -----------------------------
# Loss Computation
# -----------------------------
loss = compute_loss(detections, masks, b, p, h, w, c)
print(f"Computed symbolic loss: {loss}")

# -----------------------------
# Backpropagation Simulation
# -----------------------------
grads = backpropagation(b, L, h, w, d)

# -----------------------------
# Summary (Symbolic Output)
# -----------------------------
print("Real-image modular pipeline execution successful.")
print(f"Detections for batch 0: {detections[0]}")
print(f"Mask pixel sample for batch 0, pixel [0,0]: {masks[0][0][0]}")
```

```python
# =========================================
# Automatic Complexity Extraction
# =========================================

import json
import os
from complexity_extractor import ComplexityAnalyzer

analyzer = ComplexityAnalyzer("vision_pipeline.py")
report = analyzer.analyze()

print("\n" + "="*75)
print("        AUTOMATIC COMPLEXITY ANALYSIS REPORT")
print("="*75)

# ----------------------------
# Metadata
# ----------------------------
print("\n[Metadata]")
for key, value in report.get("metadata", {}).items():
    print(f"  {key}: {value}")

# ----------------------------
# Input Symbols
# ----------------------------
print("\n[Input Symbols]")
for key, value in report.get("input_symbols", {}).items():
    print(f"  {key}: {value}")

# ----------------------------
# Assumptions
# ----------------------------
print("\n[Assumptions]")
for assumption in report.get("assumptions", []):
    print(f"  • {assumption}")

# ----------------------------
# Components
# ----------------------------
print("\n[Pipeline Components]")
print("-"*75)

components = report.get("components", {})

for func_name, comp in components.items():
    print(f"\nFunction: {func_name}")

    time = comp.get("time", {})
    space = comp.get("space", {})

    print("  Time Complexity:")
    print(f"     O      : {time.get('O', 'N/A')}")
    print(f"     Ω      : {time.get('Omega', 'N/A')}")
    print(f"     Θ      : {time.get('Theta', 'N/A')}")

    print("  Space Complexity:")
    print(f"     O      : {space.get('O', 'N/A')}")
    print(f"     Ω      : {space.get('Omega', 'N/A')}")
    print(f"     Θ      : {space.get('Theta', 'N/A')}")

    print("-"*75)

print("="*75)

# Ensure reports directory exists
os.makedirs("reports", exist_ok=True)

output_path = "reports/complexity_report.json"
analyzer.export_report(output_path)

print(f"\nComplexity report saved to {output_path}")

# Optional: pretty preview
print("\n[JSON Preview]")
print(json.dumps(report, indent=4))
```
